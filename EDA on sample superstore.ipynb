CREATING LOSS DATAFRAME

```python
loss_df=superstore_df[superstore_df['Profit'] < 0]
loss_df
```

```python
loss_df.shape
```

```python
loss_df.describe()
```

```python
Total_loss=np.negative(loss_df['Profit'].sum())
print("Total loss = %.2f" %Total_loss)
```

```python
loss_df.groupby(by='Segment').sum()
```

##### * more discount leads to more loss,so,to make profit provide fewer discounts

```python
loss_df.groupby(by="Sub-Category").sum()
```

##### * we can observe that there are more loss in the "Binders category", "machine category", and "tables category"
##### * Binders category are getting the more sold, so giving even less discount may lead to high loss
##### * Therefore its better to give discounts where there is less sale so as to increase the sale of the products and normaliza discount rate.

```python
loss_df['Sub-Category'].value_counts()
```

```python
loss_df.groupby(by='City').sum().sort_values('Profit',ascending=True).head(10)
```

##### * The above list shows the top 10 cities with high loss
